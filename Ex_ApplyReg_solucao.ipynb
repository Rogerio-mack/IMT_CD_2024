# Setup Pandas and Numpy

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://github.com/Rogerio-mack/IMT_CD_2024/raw/main/data/movies.csv')
df.head()
```

# Q1.
Obtenha a quantidade de filmes dos anos 2016, 2017 e 2018.

1. Crie uma coluna com o ano extraído do título (`apply`e/ou expressão regular)
2. Aplique um `groupby`.

Qual a soma total dos títulos de 2016, 2017 e 2018?

```python
year_regex_pattern = '([(])([0-9]{4})([)]$)'
```

```python
df['year'] = df.title.str.extract(year_regex_pattern)[1]
df['year'] = df['year'].astype('int')
df.head()
```

```python
df[df.year > 2015 ].groupby('year').year.count().sum()
```

# Q2.

Obtenha a quantidade de filmes por década.

1. Crie uma coluna com a década a partir do ano (`apply` ou `map`)
2. Aplique um `groupby`.

Qual década teve mais títulos?

```python
def decade(year):
 return (year // 10) * 10

```

```python
df['decade'] = df.year.apply(decade)
df.head()
```

```python
df['decade'] = df.year.map(decade)
df.head()
```

```python
df.groupby('decade').decade.count()
```

# Q3.

Agrupe os ratings por `MovieId` agregando pela média.

Qual o rating do movie 1?

```python
ratings = pd.read_csv('https://github.com/Rogerio-mack/IMT_CD_2024/raw/main/data/ratings.csv')
ratings.head()
```

```python
ratings_mean = ratings.groupby('movieId').rating.mean().reset_index()
ratings_mean.head()
```
